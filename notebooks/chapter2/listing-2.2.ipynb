```python
import hopsworks

proj = hopsworks.login()
fs = proj.get_feature_store()
```

```python
import pandas as pd

df = pd.read_csv("../../data/ecommerce/product.csv")
df
```

```python
product_fg = fs.create_feature_group(
                        name="product_fg",
                        primary_key=["product_id"],
                        version=1,
                        description="Product details",
                        online_enabled=True
)

product_fg.save(df)
```
